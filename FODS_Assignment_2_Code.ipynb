## Function Definitions

```python
def data_preprocessing( data: pd.core.frame.DataFrame, scaling_type: str) -> pd.core.frame.DataFrame:
    assert scaling_type in ['standardization','min-max'] , "scaling_type takes values as min-max or standardization"
    if scaling_type == 'standardization':
        data = (data - data.mean())/data.std()
    elif scaling_type == 'min-max':
        data = (data - data.min()) / (data.max()-data.min())
    print("shape of the dataset:", df.shape)
    return data, df.columns
```

```python
def train_test_val_split(data: pd.core.frame.DataFrame, test_ratio: float, val_ratio: float, return_X_y: bool = False, random_state = 42):
    np.random.seed(random_state)
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data)*test_ratio)
    val_set_size = test_set_size + int(len(data)*val_ratio)
    test_indices = shuffled_indices[:test_set_size]
    val_indices = shuffled_indices[test_set_size:val_set_size]
    train_indices = shuffled_indices[val_set_size:]
    train_set , test_set, val_set = data.iloc[train_indices], data.iloc[test_indices], data.iloc[val_indices]
    train_X,train_y,test_X, test_y, val_X, val_y = np.asmatrix(train_set.iloc[:, :-1].values) , np.asmatrix(train_set.iloc[:,-1].values).T , np.asmatrix(test_set.iloc[:, :-1].values), np.asmatrix(test_set.iloc[:,-1].values).T , np.asmatrix(val_set.iloc[:, :-1].values), np.asmatrix(val_set.iloc[:,-1].values).T
    print("\n train_X shape:",train_X.shape, 
          "\n train_y shape:",train_y.shape, 
          "\n test_X shape:",test_X.shape, 
          "\n test_y shape:",test_y.shape, 
          "\n val_X shape:",val_X.shape,
          "\n val_y shape:",val_y.shape)
    if(return_X_y):
        # returns in order: train_X, train_y, test_X, test_y, val_X, val_y
        return train_X,train_y,test_X, test_y, val_X, val_y
    else:
        return train_set, test_set, val_set
```

```python
def pearson_correlation_feature_selection(dataset, threshold):
    target_index = dataset.shape[1] - 1
    col_corr = set()
    corr_matrix = dataset.corr()
    # checking with target variable , if > threshold keeping the predictor variable
    for i in range(target_index):
        if abs(corr_matrix.iloc[i,target_index]) > threshold:
            col_corr.add(i)
    #checking in between if >threshold , dropping one with least corr with target variable
    #for i in range(len(corr_matrix.columns)):
    #    for j in range(i):
    #        if i in col_corr and j in col_corr:
    #           if abs(corr_matrix.iloc[i,j]) > threshold:
    #               corr_i_target = abs(corr_matrix.iloc[i,target_index])
    #                corr_j_target = abs(corr_matrix.iloc[j,target_index])
    #                if corr_i_target > corr_j_target:
    #                     col_corr.remove(j)
    #                 else:
    #                     col_corr.remove(i)
    return col_corr
```

```python
def predict(X, w):
    if w.shape[0] != X.shape[1]:
        X = np.column_stack((np.ones(train_X.shape[0]), X))
    return np.asmatrix(X@w)

def loss(y, yhat):
    return np.sum(np.square(y-yhat))/2

def rms_error(y, yhat):
    return np.sqrt(np.sum(np.square(y-yhat))/len(y))

def closed_linear_regression(X, y):
    if X.ndim == 1:
        return np.asmatrix((X.T@y)/(X.T@X))
    return np.asmatrix(np.linalg.pinv(X.T@X)@(X.T@y))
```

```python
# Principal Component Analysis
def PCA(X, feature_set_size ):
    C = X.T @ X / X.shape[0]
    eigval, eigvec = np.linalg.eigh(C)
    return (eigval ,eigvec[:, [i for i in range(-1, ( -1*feature_set_size - 1 ),-1) ] ] )
```

```python
def capture(X):
    C = X.T @ X / X.shape[0]
    eigval = np.linalg.eigvals(C)
    total = sum(eigval)
    cap_list = {}
    for k in range(1,len(eigval)+1):
        cap_list[k] = (sum(eigval[: k]) / total)*100
    return cap_list
```

```python
# Greedy forward
def greedy_forward(X,val_X,y, val_y, feature_size , a =[]):
    if feature_size == 0:
        return None    
    min_rms = float('inf')
    train_error = float('inf')
    selected_feature = -1
    a_m = a.copy()
    for i in range(X.shape[1]):
        if i not in a:
            a_m.append(i)
            X_mod = X[:,a_m]
            w = closed_linear_regression(X_mod, y)
            val_X_mod = val_X[:,a_m]
            rms = rms_error(val_y, predict(val_X_mod, w))
            if rms < min_rms:
                selected_feature = i
                min_rms = rms
                train_error = rms_error(y, predict(X_mod, w))
            a_m.remove(i)
            
    a.append(selected_feature) 
    a_=[]
    for i in a:
        a_.append(column_names[i])
    print('Selected Feature Set :', a,'\n', 'rms error on training set :',train_error, '\n', 'rms error on validation set :',min_rms , end='\n\n')
    greedy_forward(X,val_X, y, val_y , feature_size - 1, a )
```

```python
# Greedy backward
def greedy_backward(X, val_X , y, val_y, feature_size , a =list(range(26)) ):
    if feature_size == 1:
        return None   
    min_rms = float('inf')
    train_error = float('inf')
    selected_feature = -1
    
    for i in range(X.shape[1]):
        if i in a:
            m = a.copy()
            m.remove(i)
            t = X[:,m]
            w = closed_linear_regression(t, y)
            val_t = val_X[:,m]
            rms = rms_error(val_y, predict(val_t, w))
            if rms < min_rms:
                selected_feature = i
                min_rms = rms
                train_error = rms_error(y, predict(t, w))
    a.remove(selected_feature)
    a_=[]
    for i in a:
        a_.append(column_names[i])
    print('feature_set:', a, '\n', 'rms error on training set:',train_error,'\n', 'rms error on validation set:',min_rms, end="\n\n" )
    greedy_backward(X, val_X, y, val_y , feature_size - 1, a )
```
